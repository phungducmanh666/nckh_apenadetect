```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# in data path
in_data_path="./out/manh"
# files name in path
nt_files_name=[]
bt_files_name=[]
# file name apena start with
file_name_prefix="nt_"
# data sample list
bt_df_list=[]
nt_df_list=[]
```

```python
bt_files_name=[f for f in os.listdir(in_data_path) if not f.startswith(file_name_prefix)]
nt_files_name=[f for f in os.listdir(in_data_path) if f.startswith(file_name_prefix)]

print("số mẫu bình thường: ", len(bt_files_name))
print("số mẫu ngưng thở: ", len(nt_files_name))
```

```python
def listAllDfFromPath(in_path, start_prefix):
    df_list = []
    for file_name in os.listdir(in_path):
        if not file_name.startswith(start_prefix):
            continue

        df = pd.read_csv(f"{in_path}/{file_name}")
        df_list.append(df)

    return df_list
```

```python
bt_df_list = listAllDfFromPath(in_data_path, "bt_")
nt_df_list = listAllDfFromPath(in_data_path, "nt_")

print("số df bình thường: ", len(bt_files_name))
print("số df ngưng thở: ", len(nt_files_name))
```

```python
def plot2Df(bt_df, nt_df, show_plot=True, show_detail=True):
    # các trục thở bình thường
    ax_bt = bt_df['ax']
    ay_bt = bt_df['ay']
    az_bt = bt_df['az']
    gx_bt = bt_df['gx']
    gy_bt = bt_df['gy']
    gz_bt = bt_df['gz']
    # các trục ngưng thở
    ax_nt = nt_df['ax']
    ay_nt = nt_df['ay']
    az_nt = nt_df['az']
    gx_nt = nt_df['gx']
    gy_nt = nt_df['gy']
    gz_nt = nt_df['gz']

    if show_plot:
        plt.figure()

        plt.subplot(6,2,1)
        plt.plot(ax_bt)
        plt.plot([ax_bt.mean()] * ax_bt.shape[0])
        plt.subplot(6,2,2)
        plt.plot(ax_nt, color="red")
        plt.plot([ax_nt.mean()] * ax_nt.shape[0], color="green")

        plt.subplot(6,2,3)
        plt.plot(ay_bt)
        plt.plot([ay_bt.mean()] * ay_bt.shape[0])
        plt.subplot(6,2,4)
        plt.plot(ay_nt, color="red")
        plt.plot([ay_nt.mean()] * ay_nt.shape[0], color="green")

        plt.subplot(6,2,5)
        plt.plot(az_bt)
        plt.plot([az_bt.mean()] * az_bt.shape[0])
        plt.subplot(6,2,6)
        plt.plot(az_nt, color="red")
        plt.plot([az_nt.mean()] * az_nt.shape[0], color="green")

        plt.subplot(6,2,7)
        plt.plot(gx_bt)
        plt.plot([gx_bt.mean()] * gx_bt.shape[0])
        plt.subplot(6,2,8)
        plt.plot(gx_nt, color="red")
        plt.plot([gx_nt.mean()] * gx_nt.shape[0], color="green")

        plt.subplot(6,2,9)
        plt.plot(gy_bt)
        plt.plot([gy_bt.mean()] * gy_bt.shape[0])
        plt.subplot(6,2,10)
        plt.plot(gy_nt, color="red")
        plt.plot([gy_nt.mean()] * gy_nt.shape[0], color="green")

        plt.subplot(6,2,11)
        plt.plot(gz_bt)
        plt.plot([gz_bt.mean()] * gz_bt.shape[0])
        plt.subplot(6,2,12)
        plt.plot(gz_nt, color="red")
        plt.plot([gz_nt.mean()] * gz_nt.shape[0], color="green")

        plt.show()

    if show_detail:

        print()
        print(f"ax bình thường min / ngưng thở min: {ax_bt.min()}/{ax_nt.min()}")
        print(f"ax bình thường max / ngưng thở max: {ax_bt.max()}/{ax_nt.max()}")
        print(f"ax bình thường mean / ngưng thở mean: {ax_bt.mean()}/{ax_nt.mean()}")
        print(f"ax bình thường std / ngưng thở std: {ax_bt.std()}/{ax_nt.std()}")
        print()
        print(f"ay bình thường min / ngưng thở min: {ay_bt.min()}/{ay_nt.min()}")
        print(f"ay bình thường max / ngưng thở max: {ay_bt.max()}/{ay_nt.max()}")
        print(f"ay bình thường mean / ngưng thở mean: {ay_bt.mean()}/{ay_nt.mean()}")
        print(f"ay bình thường std / ngưng thở std: {ay_bt.std()}/{ay_nt.std()}")
        print()
        print(f"az bình thường min / ngưng thở min: {az_bt.min()}/{az_nt.min()}")
        print(f"az bình thường max / ngưng thở max: {az_bt.max()}/{az_nt.max()}")
        print(f"az bình thường mean / ngưng thở mean: {az_bt.mean()}/{az_nt.mean()}")
        print(f"az bình thường std / ngưng thở std: {az_bt.std()}/{az_nt.std()}")
        print()
        print(f"gx bình thường min / ngưng thở min: {gx_bt.min()}/{gx_nt.min()}")
        print(f"gx bình thường max / ngưng thở max: {gx_bt.max()}/{gx_nt.max()}")
        print(f"gx bình thường mean / ngưng thở mean: {gx_bt.mean()}/{gx_nt.mean()}")
        print(f"gx bình thường std / ngưng thở std: {gx_bt.std()}/{gx_nt.std()}")
        print()
        print(f"gy bình thường min / ngưng thở min: {gy_bt.min()}/{gy_nt.min()}")
        print(f"gy bình thường max / ngưng thở max: {gy_bt.max()}/{gy_nt.max()}")
        print(f"gy bình thường mean / ngưng thở mean: {gy_bt.mean()}/{gy_nt.mean()}")
        print(f"gy bình thường std / ngưng thở std: {gy_bt.std()}/{gy_nt.std()}")
        print()
        print(f"gz bình thường min / ngưng thở min: {gz_bt.min()}/{gz_nt.min()}")
        print(f"gz bình thường max / ngưng thở max: {gz_bt.max()}/{gz_nt.max()}")
        print(f"gz bình thường mean / ngưng thở mean: {gz_bt.mean()}/{gz_nt.mean()}")
        print(f"gz bình thường std / ngưng thở std: {gz_bt.std()}/{gz_nt.std()}")


    return [
        [
            ax_bt.min() < ax_nt.min(),
            ax_bt.max() > ax_nt.max(),
            ax_bt.mean() > ax_nt.mean(),
            ax_bt.std() > ax_nt.std(),
        ],
        [
            ay_bt.min() < ay_nt.min(),
            ay_bt.max() > ay_nt.max(),
            ay_bt.mean() > ay_nt.mean(),
            ay_bt.std() > ay_nt.std(),
        ],
        [
            az_bt.min() < az_nt.min(),
            az_bt.max() > az_nt.max(),
            az_bt.mean() > az_nt.mean(),
            az_bt.std() > az_nt.std(),
        ],
        [
            gx_bt.min() < gx_nt.min(),
            gx_bt.max() > gx_nt.max(),
            gx_bt.mean() > gx_nt.mean(),
            gx_bt.std() > gx_nt.std(),
        ],
        [
            gy_bt.min() < gy_nt.min(),
            gy_bt.max() > gy_nt.max(),
            gy_bt.mean() > gy_nt.mean(),
            gy_bt.std() > gy_nt.std(),
        ],
        [
            gz_bt.min() < gz_nt.min(),
            gz_bt.max() > gz_nt.max(),
            gz_bt.mean() > gz_nt.mean(),
            gz_bt.std() > gz_nt.std(),
        ],
    ]

```
